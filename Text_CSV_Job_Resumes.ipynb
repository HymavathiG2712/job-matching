```python
import os
import pandas as pd

#Resume Corpus Path here
folder_path = ''

designation = []
for file_name in os.listdir(folder_path):
    if file_name.endswith('.lab'):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r') as file:
            file_contents = file.read()
            designation.append(file_contents)

resume_description = []

for file_name in os.listdir(folder_path):
    if file_name.endswith('.txt'):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r',encoding='ISO-8859-1') as file:
            file_contents = file.read()
            resume_description.append(file_contents)

df = pd.DataFrame({'Job_Designation': designation, 'Resume_Description': resume_description})
df
```

```python
#df.to_csv
df.to_csv("job_resumes.csv")
```
