```python
import pandas as pd
```

```python
df = pd.read_csv("last_df.csv")
```

```python
df
```

```python
import pandas as pd
import gensim
from gensim.models import Word2Vec


# Concatenate Resume_Text and Job_Description columns into a single list of sentences
sentences = [sentence for sentence in df['Resume_Text']] + [sentence for sentence in df['Job_Description']]

# Preprocess sentences by tokenizing and converting all text to lowercase
preprocessed_sentences = [
    [word.lower() for word in gensim.utils.simple_preprocess(str(sentence))]
    for sentence in sentences
]
```

```python

# Train word2vec model on preprocessed sentences
model = Word2Vec(preprocessed_sentences,vector_size=100, window=5, min_count=1, workers=4)

# Save word2vec model to file
model.save("word2vec21.model")
```
